# Definición de Costo de Vehículos

Para la definición de costo de vehículos se utilizará valores reales de la gasolina y también los litros que gasta un SUV. Con estos valores podemos darle un costo en Soles sobre cuánto gastaría un camión de delivery.
image.png

Como se puede apreciar en la imagen, el valor por litro de gasolina es 4.290 PEN, averiguando también se encuentra el monto de combustible gastado por distancia recorrida de 35 km.

image.png

```python
class Car:
    def __init__(self,x,y):
        #127 litros es la capacidad de un SUV de deliveries
        self.x = x
        self.y = y


    def cost(self,meters_run):
        #35000 m x 3.78 liters
        #4.290 PEN x Litro
        liters_used = (meters_run*3.78)/35000
        return float(4.290*liters_used)

if __name__ == "__main__":
    deposit = [15,10]
    car1 = Car(2,7)
    print(deposit,car1.x,car1.y)

    meters_per_car=0
    #hallando los metros recorridos para llegar al deposito
    while car1.x != deposit[0]:
        car1.x += 1
        meters_per_car += 1

    while car1.y != deposit[1]:
        car1.y += 1
        meters_per_car += 1

    print(f'Meters traveled: {meters_per_car}')
    print(f'Total cost of vehicle: {car1.cost(meters_per_car)} PEN')
```

Una vez obtenido los resultados con una simple regla de 3 podemos hallar el costo en soles que cada camión de delivery consume de acuerdo a la distancia recorrida al momento de ir a un depósito. El código simplificado nos sirve de ejemplo para ver cómo es que la distancia recorrida por el camión afectaría en el costo.
