# Espacio de Búsqueda para la solución de VRP
-Para tener un espacio de búsqueda es necesario definir los items para el entendimiento del problema y los distintos caminos a tomar.

*   Punto actual: Inicio de nuestro camión de delivery.
*   Puntos de Distribución: Depósitos de los cuales los vehículos irán a recoger los objetos a entregar.
*   Puntos de entrega: Ubicación donde los camiones deben ir una vez ya visitado los depósitos.
*   Puntos de entrega ya visitados: Una vez dejado el objeto de entrega, marcar este punto como ya visitado para omitir este punto al próximo camión.

image.png

# Restricciones:

Para denotar los movimientos de un camión de delivery, este debe tener ciertas restricciones.


*   No puede visitar un punto de entrega sin antes haber visitado el depósito.
*   Los camiones son ilimitados.
*   Una vez esté un punto visitado, otro camión no puede hacer una entrega ahí.
*   El camino a transitar no puede ser diagonal, solo movimientos en un eje a la vez.
*   Una vez un depósito sea visitado otro camión no podrá ir ahí hasta que el otro punto también se haya vistado, cambiando la condición de este.




# Espacios de Búsqueda

Como ya definimos en las restricciones, un camión puede ir a un punto de entrega si y solo si ha pasado por un depósito antes.
En cada iteración el costo de entrega y "cuadras" recorridas será guardado para comparar el mejor recorrido posible para realizar todas las entregas.
